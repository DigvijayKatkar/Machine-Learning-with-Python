```python
#find avg of age to fill gradyear missing values
data.groupby('gradyear')['age'].mean()
```

```python
#fill missing values with avg of age
data['age'] = data.groupby('gradyear').transform(lambda x : x.fillna(x.mean()))
```

```python
#check for missing values again
data.isnull().sum()
```

```python
#Now we will find outliers 
```

The original age range contains value from 3 - 106, which is unrealistic because student at age of 3 or 106 would not attend high school. A reasonable age range for people attending high school will be the age range between 13 to 21. The rest should be treated as outliers keeping the age of student going to high school in mind. Let's detect the outliers using a box plot below

```python
#plotting boxplot to find out outliers
sns.boxplot(data['age'])
```

```python
#finding interquntile range
q1 = data['age'].quantile(0.25)
q3 = data['age'].quantile(0.75)
iqr = q3-q1
print(iqr)
```

```python
df = data[(data['age'] > (q1 - 1.5*iqr)) & (data['age'] < (q3 + 1.5*iqr))]
```

```python
df['age'].describe()
```

From above summary you can see that the minimum age is 13.71 and maximum age is 21.15 now.

```python
#see how much observations are remaining after treating outliers
df.shape
```

```python
#check outliers using boxplot
sns.boxplot(df['age'])
```

```python
#now you have to do data preprocessing
names = df.columns[5:40]
scaled_feature = data.copy()
names
```

```python
features = scaled_feature[names]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(features.values)
```

```python
features = scaler.transform(features.values)
scaled_feature[names] = features
scaled_feature.head()
```

```python
#convert object variables to numeric variables
def gender_to_numeric(x):
    if x=='M':
        return 1
    if x=='F':
        return 2
    if x=='Not mentioned':
        return 3
```

```python
scaled_feature['gender'] = scaled_feature['gender'].apply(gender_to_numeric)
scaled_feature['gender'].head()
```

```python
#now check transformed values
scaled_feature.head()
```

```python
#now build K-means model
from sklearn.cluster import KMeans
kmeans=kmeans = KMeans(n_clusters=5, random_state=0, n_jobs=-1)
```

```python
#fit model on scaled_data
model=kmeans.fit(scaled_feature)
```

```python
#use elbow method to find out optimal number of clusters.
wcss = []
for i in range(1,20):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(scaled_feature)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,20),wcss)
plt.title('The Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

From above elbow-curve you can find out the optimal number of cluster =5

```python
#now fit kmeans for 5 clusters
kmeans = KMeans(n_clusters=5)
kmeans.fit(scaled_feature)
```

```python
#calculate the no of observations in kmeans
len(kmeans.labels_)
```

```python
data['cluster']=kmeans.labels_
data.head()
#cluster variable shows the observation belongs to which cluster
```

```python
#visualize clustering results
#show cluster size
plt.figure(figsize=(12,7))
axis = sns.barplot(x=np.arange(0,5,1),y=data.groupby(['cluster']).count()['age'].values)
x=axis.set_xlabel("Cluster Number")
x=axis.set_ylabel("Number of Students")
```

From above plot you can say that the cluster 2 has largest size and cluster 3 has smallest.

```python
#check number of observations belonging to each cluster
size_array = list(data.groupby(['cluster']).count()['age'].values)
size_array
```
